```python
# create an algorithm that displays all numbers from 1 to n, color in its divisors and tell if it is prime or not
n=int(input("\nenter a number: "))
s=0

# extracting the data and working the print
for i in range(1,n+1):
    if n%i==0:
        print(F"\033[1;32m{i}\033[m")
        s+=1
    else:
        print(f"\033[1;34m{i}\033[m")
        
# print if it's a prime number
if s>2:
    print("it's not a prime number")
else:
    print("it's a prime number")
```
